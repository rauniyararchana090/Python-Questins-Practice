```python
import re 
def convert_into_uppercase(a):
    return a.group(1) + a.group(2).upper()
s ="hello world I am archana rauniyar"
result = re.sub("(^|\s)(\S)", convert_into_uppercase, s)
result= result.replace(" ","")
print(result)
```
